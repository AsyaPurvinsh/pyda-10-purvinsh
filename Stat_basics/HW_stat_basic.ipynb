```python
import pandas as pd
import numpy as np
```

```python
data = data.replace('?',np.NaN)
data
```

```python
#посмотрим выбросы по числовым значениям
#по температуре
q1_t = data['rectal temperature'].quantile(0.25)
q3_t = data['rectal temperature'].quantile(0.75)
iqr_t = q3_t - q1_t
lower_bound_t = q1_t - (1.5 * iqr_t) 
upper_bound_t = q3_t + (1.5 * iqr_t)
lower_outliers_t = data[data['rectal temperature']<lower_bound_t]
upper_outliers_t = data[data['rectal temperature']>upper_bound_t]
```

```python
lower_outliers_t 
```

Выбросы не какие-то специфичные, появлились естественным путем, оставляем их


```python
upper_outliers_t
```

Верхние выбросы по температуре тоже не кажутся ошибочными, тем более большая часть лошадей болеет или в стрессе. Также нормальность данных подтверждает высокий пульс. Ничего с выбросами не делаем

```python
#по пульсу
q1_p = data['pulse'].quantile(0.25)
q3_p = data['pulse'].quantile(0.75)
iqr_p = q3_p - q1_p
lower_bound_p = q1_p - (1.5 * iqr_p) 
upper_bound_p = q3_p + (1.5 * iqr_p)
lower_outliers_p = data[data['pulse']<lower_bound_p]
upper_outliers_p = data[data['pulse']>upper_bound_p]
```

Нижних выбросов нет

```python
upper_outliers_p
```

Верхние выбросы появились естественным путем - от измерения у лошадей с сильно повышенной температурой и почти все после операции 

```python
#по частоте дыхания
q1_r = data['respiratory rate'].quantile(0.25)
q3_r = data['respiratory rate'].quantile(0.75)
iqr_r = q3_r - q1_r
lower_bound_r = q1_r - (1.5 * iqr_r) 
upper_bound_r = q3_r + (1.5 * iqr_r)
lower_outliers_r = data[data['respiratory rate']<lower_bound_r]
upper_outliers_r = data[data['respiratory rate']>upper_bound_r]
```

```python
upper_outliers_r
```

По частоте дыхания ткакже нет нижних выбросов и много верхних с также высокими показателями температуры и пульса

Итого: все выбросы в выборке появились естественным путем в виду болезни лошадок - принимаем решение ничего с нимине делать и полностью учитывать при расчетах

```python
#проверим кол-во пропусков данных
for col in data.columns:
    pct_missing = data[col].isnull().mean()
    print(f'{col} - {pct_missing :.1%}')
```

```python
#Для начала удалим полностью строки, где заполненно меньше половины значений 
data = data.dropna(thresh=4)
```

Осталось 291 строка

```python
#проверим еще раз кол-во пропусков данных
for col in data.columns:
    pct_missing = data[col].isnull().mean()
    print(f'{col} - {pct_missing :.1%}')
```

```python
data[data.Surgery.isnull()]
```

```python
#посмотрим в каких случаях дедали операцию чаще
data.groupby('Surgery')[['rectal temperature','pulse', 'respiratory rate']].median()
```

После операции показатели пульса выше, поэтому предположим что в данном случае с пропуском операцию лошадке не делали

```python
data['Surgery'].fillna('2', inplace=True)
```

```python
#пропуски в данных о ректальной температуре
len(data[data['rectal temperature'].isnull()])
```

```python
#возьмем медианные значения при разных значениях операции и температуры конечностей 
data.groupby(['Surgery', 'temperature of extremities'])['rectal temperature'].median()
```

```python
#заполним пропуски этими значениями
data['rectal temperature'].fillna(data.groupby(['Surgery', 'temperature of extremities'])['rectal temperature'].transform('median'), inplace=True)
```

```python
data[data['rectal temperature'].isnull()]
```

Не все данные заполнились, остались те, в которых температура конечностей не заполнена
Так как планируется заполнять пропуски в температуре конечностей основываясь в том числе на ректальной температуре, то эти  данные стоит удалить, чтобы не добавить искажения результатов)

```python
data = data.dropna(subset=['rectal temperature'])
```

```python
#пропуски в данных о пульсе
data[data.pulse.isnull()]
```

```python
#заполним  пропуски данными исходя из медианных значений переферийного пульса
print(data.groupby('peripheral pulse')['pulse'].median())
data['pulse'].fillna(data.groupby('peripheral pulse')['pulse'].transform('median'), inplace=True)
```

```python
#пропуски в данных о частоте дыхания
len(data[data['respiratory rate'].isnull()])
```

```python
#добавим новые столбец с оценкой пульса, разделив его на 4 категории
data.loc[data['pulse'] < 49, 'pulse_rate'] = 'low'
data.loc[(data['pulse'] > 48) & (data['pulse'] < 65), 'pulse_rate'] = 'normal'
data.loc[(data['pulse'] > 64) & (data['pulse'] < 89), 'pulse_rate'] = 'increased'
data.loc[data['pulse'] > 88, 'pulse_rate'] = 'high'
```

```python
#заполним пропуски в данных о частоте дыхания медианными значения при разных значениях операции и оценки пульса
print(data.groupby(['Surgery', 'pulse_rate'])['respiratory rate'].median())
data['respiratory rate'].fillna(data.groupby(['Surgery', 'pulse_rate'])['respiratory rate'].transform('median'), inplace=True)
```

```python
#пропуски в данных о температуре конечностей
len(data[data['temperature of extremities'].isnull()])
```

```python
#также создадим доп столбец с оценкой ректальной температуры
data.loc[data['rectal temperature'] < 37.9, 'rectal temperature_rate'] = 'low'
data.loc[(data['rectal temperature'] > 37.8) & (data['rectal temperature'] < 38.3), 'rectal temperature_rate'] = 'normal'
data.loc[(data['rectal temperature'] > 38.2) & (data['rectal temperature'] < 39), 'rectal temperature_rate'] = 'increased'
data.loc[data['rectal temperature'] > 39.1, 'rectal temperature_rate'] = 'high'
```

```python
#заполним пропуски данными исходя из модальных значений оценки ректальной температуры
data.groupby('rectal temperature_rate')['temperature of extremities'].agg(pd.Series.mode)
```

При любой температуре модальное значение температуры конечностей = 3. Заполним все пропуски этим значением

```python
data['temperature of extremities'].fillna('3', inplace=True)
```

```python
#пропуски в данных о переферическом пульсе
len(data[data['peripheral pulse'].isnull()])
```

```python
#заполним пропуски данными исходя из модальных значений оценки пульса
print(data.groupby('pulse_rate')['peripheral pulse'].agg(pd.Series.mode))
f = lambda x: x.mode().iloc[0]
data['peripheral pulse'].fillna(data.groupby('pulse_rate')['peripheral pulse'].transform(f), inplace=True)
```

```python
len(data[data['mucous membranes'].isnull()])
```

```python
#заполним пропуски данными исходя из модальных значений оценки пульса и переферического пульса
print(data.groupby('pulse_rate')['mucous membranes'].agg(pd.Series.mode))
f1 = lambda x: x.mode().iloc[0]
data['mucous membranes'].fillna(data.groupby('pulse_rate')['mucous membranes'].transform(f1), inplace=True)
```

```python
#проверим, что пропусков не осталось
for col in data.columns:
    pct_missing = data[col].isnull().mean()
    print(f'{col} - {pct_missing :.1%}')
```
